```python
# Just removing the brackets

marvel['active_years'] = marvel['active_years'].apply(lambda x: x[1:-1])    
```

#### Search for Spiderman and other comics

```python
import re
```

```python
def find_spiderman(comic):
    x = re.search("..Spider..", comic)
    if x:
        return comic
    else:
        return 'None'
```

```python
spiderman_comics = marvel.copy()
```

```python
spiderman_comics['comic_name'] = marvel['comic_name'].apply(lambda x: find_spiderman(x))
```

```python
spiderman_comics[spiderman_comics['comic_name'] == 'None']
```

```python
spiderman_comics = spiderman_comics[spiderman_comics['comic_name'] != 'None']
```

```python
spiderman_comics.head()
```

#### Turn price into numeric

```python
spiderman_comics['Price'].value_counts()
```

```python
spiderman_comics[spiderman_comics['Price'] == 'None']
```

```python
def convert_to_float(price):
    if price.strip() == 'Free':
        return 0.
    if price.strip() == 'None':
        return np.NaN
    else:
        num = price.strip()[1:]
        return float(num)
```

```python
spiderman_comics['Price'] = spiderman_comics['Price'].apply(lambda x: convert_to_float(x))
```

```python
spiderman_comics['Price'].head()
```

```python
spiderman_comics['Price'].plot(kind='hist', bins=40, figsize=(12, 8))

# We can see that most of the spiderman comics have a price
# between 0 (Free) and 1$
```
