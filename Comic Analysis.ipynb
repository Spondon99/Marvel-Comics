## Marvel comics analysis

### Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Read the file and briefly analyze it

```python
marvel = pd.read_csv('Marvel_Comics.csv')
```

```python
marvel.head()
```

```python
marvel.info()
```

```python
# Just removing the brackets

marvel['active_years'] = marvel['active_years'].apply(lambda x: x[1:-1])    
```

### Search for Spiderman comics

```python
import re
```

```python
def find_word(word, target):
    x = re.search(f"..{target}..", word)
    if x:
        return word
    else:
        return 'None'
```

```python
spiderman_comics = marvel.copy()
```

```python
spiderman_comics['comic_name'] = marvel['comic_name'].apply(lambda x: find_word(x, 'Spider'))
```

```python
spiderman_comics[spiderman_comics['comic_name'] == 'None']
```

```python
spiderman_comics = spiderman_comics[spiderman_comics['comic_name'] != 'None']
```

```python
spiderman_comics.head()
```

#### Turn price into numeric

```python
spiderman_comics['Price'].value_counts()
```

```python
spiderman_comics[spiderman_comics['Price'] == 'None']
```

```python
def convert_to_float(price):
    if price.strip() == 'Free':
        return 0.
    if price.strip() == 'None':
        return np.NaN
    else:
        num = price.strip()[1:]
        return float(num)
```

```python
spiderman_comics['Price'] = spiderman_comics['Price'].apply(lambda x: convert_to_float(x))
```

```python
spiderman_comics['Price'].head()
```

```python
spiderman_comics['Price'].plot(kind='hist', bins=40, figsize=(12, 8))

# We can see that most of the spiderman comics have a price
# between 0 (Free) and 1$
```

### Now let's see Daredevil comics

```python
daredevil_comics = marvel.copy()
```

```python
daredevil_comics['comic_name'] = marvel['comic_name'].apply(lambda x: find_word(x, 'Dare'))
```

```python
daredevil_comics.head()
```

```python
daredevil_comics = daredevil_comics[daredevil_comics['comic_name'] != 'None']
```

```python
daredevil_comics.head()
```

##### Let's create a column for starting year of the comic

```python
daredevil_comics['start_year'] = daredevil_comics['active_years'].apply(lambda x: x[:4])
```

```python
daredevil_comics.head()
```

##### Convert the start_year to int

```python
daredevil_comics['start_year'] = daredevil_comics['start_year'].astype('int16')
```

```python
daredevil_comics.describe()
```

##### Visualize the timeline of the comics' first publication

```python
sns.histplot(
    daredevil_comics['start_year'], 
    bins=(daredevil_comics['start_year'].max() - daredevil_comics['start_year'].min())
    )
```
